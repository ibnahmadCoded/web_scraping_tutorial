```python
# Get some of the author 
for div in new_soup.findAll('div', attrs={'class':'product-info-list'}):
    spans = new_soup('span')
    for i, span in enumerate(spans):
        if span.text == 'Yazar: ':
            author = spans[i + 1].text
            
    authors.append(author)
```

```python
# Get some of the book name info
for h1 in new_soup.findAll('h1', attrs={'class':'mt0'}):
    
    name = " ".join(h1.text.split())
            
    books.append(name)
```

```python
# Get some of the publication year info
pub_yr = "Not Availaible"
for b in new_soup.findAll('b'):
    if b.text == "Bask Yl: ":
        pub_yr = b.next_sibling



if pub_yr == "Not Availaible":
    for li in new_soup.findAll('li'):
        if li.span != None:   
            if li.span.text == 'lk Bask Yl:':
                pub_yr = li.a.text
                
pub_yrs.append(pub_yr)
```

```python
# Get Price info
for div in new_soup.findAll('div', attrs={'class':'current-price'}):
    
    price = div.text
            
    prices.append(price)
```

```python
# Get Number of Reviewers' info
reviews = []
review_count = "0"
for p in new_soup.findAll('p', attrs={'id':'reviewBody'}):
    
    reviews.append(p)


if len(reviews) >= 3:
    try:
        review_count = new_soup.find('input',{'id':'ModelReviewCount'})['value']
    except:
        pass
elif len(reviews) < 3 and len(reviews) > 0:
    review_count = len(reviews)
else:
    pass

num_reviews.append(review_count)
```

```python
# Get list of all ratings
ratings = []
if int(review_count) > 0:
    a = []
    for s in new_soup.findAll("strong"):
        a.append(s.text[1:])

    
    for val in a:
        content = list(val)
        try:
            int(content[0])
            new_val = []
            for i,v in enumerate(content):
                if v == ",":
                    content[i] = "."
            res = "".join(content)
            ratings.append(res)

        except ValueError:
            continue

if len(ratings) == 5:
    ratings[0] = "5 Stars: " + ratings[0] + "%"
    ratings[1] = "4 Stars: " + ratings[1] + "%"
    ratings[2] = "3 Stars: " + ratings[2] + "%"
    ratings[3] = "2 Stars: " + ratings[3] + "%"
    ratings[4] = "1 Star: " + ratings[4] + "%"
    
    ratings = ", ".join(ratings)
elif ratings == []:
    ratings = "Not Availaible"

reviewer_rates.append(ratings) 
```

```python
# Get the number of pages info
pages = ""
for b in new_soup.findAll('b'):
    if b.text == "Sayfa Says: ":
        pages = b.next_sibling

if pages == "":
    pages = "Not Availaible"
    
num_pages.append(pages)
```

```python
df = pd.DataFrame({'Book Title':books, 'Number of Pages':num_pages, 'Author':authors,
                   'Price':prices, 'Publication Year':pub_yrs, 'Number of Reviews':num_reviews, 
                   'Rates Given By Reviewers':reviewer_rates}) 

print(df)
```
